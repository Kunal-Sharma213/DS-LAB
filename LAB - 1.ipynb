```python
import matplotlib.pyplot as plt
import random
import time
import datetime
import pandas as pd
import numpy as np
```

```python
def summation(arr , i , l):
    if(i == l):
        return 0;
    return arr[i] + summation(arr , i + 1 , l)
```

```python
def linear_search(arr, i ,l , key):
    if(i == l):
        return False
    if(arr[i] == key):
        return True
    return linear_search(arr, i+1, l, key)
```

```python
def maximum(arr, i, l, mymax):
    if(i == l):
        return mymax
    if(arr[i] > mymax):
        mymax = arr[i]
    return maximum(arr,i+1,l,mymax)
```

```python
def binary_search(arr,lb,ub,key):
    if(lb > ub):
        return False
    mid = lb + (ub - lb) // 2
    if(arr[mid] == key):
        return True
    elif(arr[mid] > key):
        return binary_search(arr,lb , mid - 1, key)
    else:
        return binary_search(arr, mid + 1, ub, key)
```

```python
def truth_table(arr,i , n):
    if(i == n):
        return True
    arr[i] = 'T'
    truth_table(arr,i+1,n)
    arr[i] = 'F'
    truth_table(arr,i+1,n)
```

```python
def power(x,y):
    if(y == 0):
        return 1
    if(y%2==0):
        return power(x*x , y /2)
    else:
        return x *power(x*x , (y - 1)/2)
```

```python
time_duration = []
for i in range(1,20):
    arr = ['' for _ in range(i)]
    start = datetime.datetime.now()
    truth_table(arr, 0 , len(arr))
    end = datetime.datetime.now()
    diff = end - start
    time_duration.append(diff.total_seconds() * 1000000)
```

```python
dic = {'duration':time_duration}
```

```python
df = pd.DataFrame(dic)
```

```python
plt.figure(figsize = (20,10))
plt.plot(df)
```
